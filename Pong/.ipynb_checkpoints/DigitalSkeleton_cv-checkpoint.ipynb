```python
import numpy as np
import cv2
from cvzone.PoseModule import PoseDetector

def pad_with(vector, pad_width, iaxis, kwargs):
    pad_value = kwargs.get('padder', 255)
    vector[:pad_width[0]] = pad_value
    vector[-pad_width[1]:] = pad_value

cap = cv2.VideoCapture(0)
detector = PoseDetector(trackCon = 0.7)
width = 480
height = 640
padding = 200
r_hand = 0
l_hand = 0
while True:
    success, img = cap.read()
    copy = np.zeros((width+2*padding,height+2*padding,3),dtype=np.uint8)
    for i in range(3):
        copy[:,:,i] = np.pad(img[:,:,i], padding, pad_with)
    img = copy
    img = detector.findPose(img)
    lmlist, bboxInfo = detector.findPosition(img)
    try:
        b = bboxInfo["bbox"]
        upper_bounder = lmlist[5][2]
        lower_bounder = lmlist[12][2] + height/2
        r_count = 0
        l_count = 0
        diff = lower_bounder-upper_bounder

        for i in [16,22,18,20]:
            if len(lmlist) == 33:
                r_hand += lmlist[i][2]
                r_count += 1

        try:
            r_hand/=r_count
        except:
            r_hand = 0
        r = lower_bounder if r_hand>lower_bounder else r_hand
        r = ((upper_bounder if r<upper_bounder else r)-upper_bounder)/diff

        for i in [15,21,17,19]:
            if len(lmlist) == 33:
                l_hand += lmlist[i][2]
                l_count += 1
        try:
            l_hand/=l_count
        except:
            l_hand = 0
        l = lower_bounder if l_hand>lower_bounder else l_hand
        l = ((upper_bounder if l<upper_bounder else l)-upper_bounder)/diff

    except:
        l,r = 0.5,0.5
    with open("inputs.txt","r+") as f:
        f.seek(0)
        f.write(f"{l},{r}")
        f.truncate()
        
    cv2.imshow("Image",img)
    
    if cv2.waitKey(3) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
